```python
#Imports
import pandas as pd
import numpy as np
import geocoder
```

```python
# Get the data from Part 1
df_neighbourhoods = pd.read_csv('Toronto.csv',index_col=[0])
df_neighbourhoods.head()
```

```python
#Get the coordinates from a csv file
url_csv = 'http://cocl.us/Geospatial_data'
df_coordinates = pd.read_csv(url_csv)
df_coordinates.head()
```

```python
#Put the same name on Postcode column
df_coordinates.rename(columns={'Postal Code': 'Postcode'}, inplace=True)
df_neighbourhoods.rename(columns={'Postcode': 'Postcode'}, inplace=True)
```

```python
#Merge the two dataframes
df = pd.merge(df_neighbourhoods, df_coordinates, on='Postcode')
df.head()
```

```python
#Create a test dataframe with the same entries on the example of the assignment for testing
column_names = ["Postcode", "Borough", "Neighbourhood",'Latitude','Longitude']
test_df = pd.DataFrame(columns=column_names)

test_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in test_list:
    test_df = test_df.append(df[df["Postcode"]==postcode], ignore_index=True)
    
test_df
```

```python
#Save to a CSV file
df.to_csv('Toronto2.csv')
```
