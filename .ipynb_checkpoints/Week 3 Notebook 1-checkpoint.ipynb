```python
#Imports
import pandas as pd
import numpy as np
import requests
```

```python
#Get the url
url  = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = requests.get(url)
if page.status_code == 200:
    print('Page download successful')
else:
    print('Page download error. Error code: {}'.format(page.status_code))
```

```python
#Create a dataframe
df = pd.read_html(url, header=0, na_values = ['Not assigned'])[0]
df.head()
```

```python
#Exclude NaN Values
df.dropna(subset=['Borough'], inplace=True)
```

```python
#See the Neighbourhoods with Nan Values
df[df['Neighbourhood'].isna()]
```

```python
#Change the NaN Values with the name of the Borough
df['Neighbourhood'].fillna(df['Borough'], inplace=True)
n_empty_neighborhood = df[df['Neighbourhood'].isna()].shape[0]
df[df['Borough']=="Queen's Park"]
```

```python
df.head(10)
```

```python
#Group the Neighbourhood with same Postcode
df = df.groupby(['Postcode','Borough']).Neighbourhood.agg([('Neighbourhood', ', '.join)])
df.reset_index(inplace=True)
df.head(10)
```

```python
#Create a test dataframe with the same entries on the example of the assignment for testing
column_names = ["Postcode", "Borough", "Neighbourhood"]
test_df = pd.DataFrame(columns=column_names)

test_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in test_list:
    test_df = test_df.append(df[df["Postcode"]==postcode], ignore_index=True)
    
test_df
```
